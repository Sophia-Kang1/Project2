```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

**Columns**
* age(나이) : 20 ~ 64세
* gender(성별) : M - male / F - Female
* height_cm(키)
* weight_kg(몸무게)
* body fat_%(체지방률)
* diastolic : 이완기 혈압(min)
* systolic : 수축기 혈압(min)
* gripForce : 악력(kg)
* sit and bend forward_cm : 앉아서 앞으로 구부리기(좌전굴)

```python
# 이상치 제거
outlier_fat = df[(df['body fat_%'] >= 50)].index
outlier_bend = df[(df['sit and bend forward_cm'] >= 40)].index
df = df.drop(outlier_fat|outlier_bend,axis=0)
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 'age' 컬럼 데이터 형식 변경
df = df.astype({'age':'int'})
```

```python
# BMI 컬럼 생성
df["BMI"] = (df["weight_kg"] / ((df['height_cm']/100)**2)).round(1)
df
```

```python
Obesity_BMI = []
```

```python
# BMI별 비만도
'''
underweight : 저체중
normal : 정상
pre-obese : 비만 전단계
obese_1 : 1단계 비만
obese_2 : 2단계 비만
obese_3 : 3단계 비만
'''
for row in df['BMI']:
  if row >= 35:
    Obesity_BMI.append('obese_3')
  elif 30 <= row < 35:
    Obesity_BMI.append('obese_2')
  elif 25 <= row < 30:
    Obesity_BMI.append('obese_1')
  elif 23 <= row < 25:
    Obesity_BMI.append('pre-obese')
  elif row < 18.5:
    Obesity_BMI.append('underweight')
  else:
    Obesity_BMI.append('normal')
```

```python
df['Obesity_BMI'] = Obesity_BMI
df
```

```python
# 근감소증 조건식 작성
sarcopenia = ((df.gender == 'M') & (df.gripForce <= 28))|((df.gender == 'F') & (df.gripForce <= 18))
df['Sarcopenia'] = sarcopenia
```

```python
# 학습 / 평가 데이터셋을 분리합니다.
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# XGBoost 모델
from category_encoders import OrdinalEncoder
from sklearn.impute import SimpleImputer
from xgboost import XGBClassifier
from sklearn.pipeline import make_pipeline


pipe_xgb = make_pipeline(
    OrdinalEncoder(),
    SimpleImputer(strategy="median"),
    XGBClassifier(
        objective="multi:softmax",
        eval_metric="error",
        n_estimators=200,
        random_state=42,
        n_jobs=-1,
        max_depth=7,
        learning_rate=0.1,
    ),
)

pipe_xgb.fit(X_train, y_train)
print("훈련 정확도: ", pipe_xgb.score(X_train, y_train))
print(classification_report(y_test, pipe_xgb.predict(X_test)))
```

```python
# 모델 재학습
pipe_xgb = make_pipeline(
    OrdinalEncoder(),
    SimpleImputer(strategy="median"),
    XGBClassifier(
        objective="multi:softmax",
        eval_metric="error",
        n_estimators=200,
        random_state=42,
        n_jobs=-1,
        max_depth=6,
        learning_rate=0.1,
        min_child_weight= 8,
    ),
)

pipe_xgb.fit(X_train, y_train)
print("훈련 정확도: ", pipe_xgb.score(X_train, y_train))
print(classification_report(y_test, pipe_xgb.predict(X_test)))
```

```python
features = df.drop(columns=target)
score = pipe_xgb.score(X_test, y_test)

pi = pd.Series(dtype=float)

n_iter = 10  # 랜덤으로 특성을 섞어 평가 성능을 확인하는 과정을 몇 번 반복할지 결정합니다.

for feature in features:
    X_test_permed = X_test.copy()

    # 모델을 재학습하지 않고, 기존 모델에서 feature 값을 여러 번(n_iter) 섞고 검증하여 평균 성능을 확인합니다.
    scores_permutated = []
    for _ in range(n_iter):
        X_test_permed[feature] = np.random.permutation(X_test_permed[feature])
        scores_permutated.append(pipe_xgb.score(X_test_permed, y_test))
    avg_score_permutated = np.mean(scores_permutated)
    print(f"Perm {feature} :: 검증 정확도 {avg_score_permutated:.6f}")

    # permutation importance는 해당 특성을 랜덤으로 섞었을 때의 성능 하락폭으로 정의됩니다.
    pi[feature] = score - avg_score_permutated
```

```python
pi.sort_values().plot.barh()
plt.title("Permutation Importance")
```
